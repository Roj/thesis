```python
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['figure.dpi'] = 150
```

```python
import numpy as np
X = np.linspace(0, 200, num=200+1)
plt.plot(X, [1-0.995**x for x in X], c="#ffa600", label="p=0.05")
plt.plot(X, [1-0.99**x for x in X], c="#ff6361", label="p=0.1")
plt.legend()
plt.vlines(150, ymin=0, ymax=1)
plt.xlabel("Cantidad de alumnos")
plt.ylabel("probabilidad de contagio\n[asumiendo incubadora perfecta]")
plt.title("Porcentaje de cat√°strofe")
```
