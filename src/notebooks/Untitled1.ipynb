```python
def fnv(s, offset=14695981039346656037, prime=1099511628211):
    h = offset
    for char in s:
        h = h * prime
        h = h ^ ord(char)
    return h
```

```python
with open("/tmp/ficciones.txt") as f:
    texto = f.read().replace("\n", " ").lower()
print(f"Longitud del texto: {len(texto)}")
palabras = set(texto.split(" "))
print(f"Cantidad de palabras (aprox.): {len(palabras)}")
```

```python
N = 101 # buckets
contadores = [0]*N
for palabra in palabras:
    contadores[fnv(palabra)%N] += 1
```

```python
import matplotlib.pyplot as plt
plt.figure(dpi=200)
plt.bar(range(N), contadores)
```

```python
N = 101 # buckets
contadores = [0]*N
for palabra in palabras:
    contadores[fnv(palabra, 1024, 255)%N] += 1

plt.figure(dpi=200)
plt.bar(range(N), contadores)
```
